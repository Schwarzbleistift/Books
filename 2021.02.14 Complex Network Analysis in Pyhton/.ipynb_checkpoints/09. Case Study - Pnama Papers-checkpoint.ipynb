```python
import csv
import pickle
import itertools
from collections import Counter
import networkx as nx
from networkx.drawing.nx_agraph import graphviz_layout
import matplotlib.pyplot as plt
```

```python
EDGES = "beneficiary"
NODES = (("bahamas_leaks.nodes.entity.csv", "jurisdiction", "name"),
("bahamas_leaks.nodes.officer.csv", "country_codes", "name"),
("bahamas_leaks.nodes.intermediary.csv", "country_codes", "name"))
```

```python
Panama = nx.Graph()
with open(r'E:\OneDrive - 同济大学\05 - 数据集\0 - 统计数据集\Panama papers\bahamas_leaks.edges.csv') as infile:
    rows = csv.DictReader(infile)
    Panama.add_edges_from((row['node_1'], row['node_2']) for row in rows if row['rel_type'] == 'similar_company_as')
```

```python
nodes  = set(Panama.nodes())
relabel = {}

for file, cc, name in NODES:
    with open('E:\\OneDrive - 同济大学\\05 - 数据集\\0 - 统计数据集\\Panama papers\\' + file) as infile:
        kind = file.split('.')[-2]
        rows = csv.DictReader(infile)
        names_countries = {}
        for row in rows:
            if row['node_id'] in nodes:
                names_countries[row['node_id']] =  (row[name].strip().upper(), row[cc])
    names   =  {nid: values[0] for nid, values in names_countries.items()}
    countries ={nid: values[1] for nid, values in names_countries.items()}
    kinds   =  {nid: kind      for nid, _      in names_countries.items()}
    nx.set_node_attributes(Panama, 'country', countries)
    nx.set_node_attributes(Panama, 'kind', kinds)
    relabel.update(names)

nx.relabel_nodes(Panama, relabel, copy = False)
```
