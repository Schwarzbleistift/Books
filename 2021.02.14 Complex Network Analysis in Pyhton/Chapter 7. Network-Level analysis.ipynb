## Components/Isolates
1. **Component**: A set of actors that are internally connected but are disconnected from the rest of vertices in the network.
2. **Isolte**: When a vertex is isolated from other vertices, it simply becomes isolate.

## Core/Periphery
1. In **network cores**, members are denstly tied to each othter.
2. In a **periphery**, nodes are more linked to the core members than to each other
> It is found that actors with many ties (actors at the network core) and actors with only a few ties (actors at the periphery) have constrained and predictable behaviral patterns. Yet, actors with only some ties do not have predictable behaviors. 

## Density
1. **Density** is defined as the number of network edges divided by the max possible nuber of edges between nodes in that network.

```python
import networkx as nx
%matplotlib inline
import matplotlib.pyplot as plt
from networkx.algorithms import bipartite

g = nx.karate_club_graph()
nx.draw(g)
nx.density(g)
```

## Shortest Path
1. **Diameter** or **Max Eccentricity**: the longest distance in a graph. The goal of network diameter measure is to index the extensiveness of the network.
2. **Radius**: the min eccentricity of the network.

```python
print(nx.shortest_path(g,8,5))
print(nx.shortest_path_length(g,0,8))
print(nx.average_shortest_path_length(g))
for path in nx.all_pairs_shortest_path(g):
    print(path)
```

```python
print(nx.eccentricity(g))
print(nx.diameter(g))
print(max(nx.eccentricity(g).values()))
print(nx.radius(g))
print(min(nx.eccentricity(g).values()))
```

## Reciprocity
1. **Reciprocity**: the ratio of the number of edges (relationships) pointing in both directions to the total number of relationships in the graph (or social network).

## Affiliation networks
1. **Affiliation networks**: networks that consist of at least 2 different sets of nodes (called actors and events) in a way that ties (also called affiliations) connect nodes from differernt sets only.They are called affiliation networks to express node multi-membership or multi-participation.

2. **Two-mode network**: affiliation graphs are sometimes called two-mode, bipartite, or bimodel network. They are represented by bipartite graphs B = (V1, V2, E), where V1 and V2 are the two different sets of nodes, while E is a set of edges that only connect nodes from opposite sets.
  
  Some network measures (such as completeness and density) should be computed for two-mode networks in a way that is distinct from one-mode networks. E.g., completeness and density are usually much higher in one-mode network than in two-mode network.
  
  It is possible to change the two-mode network into a one-mode network. This is done by dividing the nodes into two disjoint sets such that edges are no more connecting nodes from different sets. This results in two networks of one mode, and thus standard analysis techniques can be applied. In the following graph, the two sets of nodes, V1 and V2, were connected through a set of nodes, E. Edges are connecting between nodes of the same set rather than of different sets. The establishment of the new edges if done based on node co-membership or co-participation in the original network.In many cases, the derived network is not simple, which means it may contain loops.
  

## Homophily
1. Homophily: One of the benefits obtained when deriving one-mode networks from a two-mode network is the identification of interaction patterns, at least more often than with dissimilar people, between similar people.


```python
# in this example, we r going to generate a very simple bipartite graph
B = nx.Graph()
B.add_nodes_from([1,2,3,4,5,6,7,8], bipartite=0)
B.add_nodes_from(['a','b','c','d','e'], bipartite=1)
B.add_edges_from([(1,'a'), (1,'b'), (2,'b'), (2,'c'), (3,'c'),(3,'d'), (4, 'd'), (4, 'e'),(5, 'd'), (5, 'e'), (6, 'd'), (7,'d'),(8, 'e'), (8, 'a')])
bottom_nodes, top_nodes = bipartite.sets(B)
print('whether it is bipartite: ' + str(bipartite.is_bipartite(B)))
print('whether it is connected: ' + str(nx.is_connected(B)))
print(bipartite.density(B, bottom_nodes))
nx.draw_networkx(B, with_labels = True)
```

```python
#in this example, we continue our work from the previous one by converting the bipartite graph into two one_mode networks.
#The bipartite.projected_graph() function returns the projection of B on one of its node sets
To = bipartite.projected_graph(B, top_nodes)
nx.draw_networkx(To)
Bo = bipartite.projected_graph(B, bottom_nodes)
nx.draw_networkx(Bo)
```

```python
#In this example, we are going to create complete bipartite graphs straightly
B = nx.complete_bipartite_graph(3,2)
nx.draw(B)
for club in bipartite.sets(B):
    print(bipartite.density(B,club))
```

```python
#In this example, we build a bipartite random graph
randB = bipartite.random_graph(15,10,0.5)
X, Y = bipartite.sets(randB)
position = dict()
position.update((n,(1,i)) for i, n in enumerate(X))
position.update((n,(2,i)) for i, n in enumerate(Y))
nx.draw(randB, pos = position, with_labels = True, node_size = 250, node_color = 'magenta')
```
